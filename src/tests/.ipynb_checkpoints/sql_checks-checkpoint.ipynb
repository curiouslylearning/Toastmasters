```python
import pandas as pd
from etl.sql_connection import sql_engine
from sqlalchemy import text
from datetime import datetime
```

```python
engine = sql_engine()
```

```python


#Parameteriserade queries

sql = text("""
SELECT *
FROM "ClubPerformance"
WHERE "TM_Year" = :tm_year
  AND "Club numeric" = :club
""")

df_club = pd.read_sql(
    sql,
    engine,
    params={
        "tm_year": "2024-2025",
        "club": 4700
    }
)

df_club
```

```python
#create a new table



def create_table_sql(table_name: str, df: pd.DataFrame):
    starttime = datetime.now()

    df.to_sql(
        #name="FindAClub",
        name=table_name,
        con=engine,
        schema="public",
        if_exists="replace", #eller append. Varning! Replace tar bort hela och lägger in ny
        index=False,
        chunksize=50_00, #5000
        #method="multi" #Innebär att den gör en enda stor Insert, alla rader på en gång typ (i chunksize) tog bort denna, hänger sig. Men ska vara effektivt ibland.
    )

    endtime = datetime.now()
    timediff = endtime - starttime
    print("Scriptet tog : " + str(timediff))



```

```python
table_name = "month_sort"

create_table_sql(table_name, df2)
```
